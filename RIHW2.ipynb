```python
import geopandas as gp
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

places = gp.read_file("/Users/romaningersoll/AES_408/cb_2024_01_place_500k/cb_2024_01_place_500k.shp")

places.plot()


```

```python
places = places.to_crs(epsg=26929)
area = places.geometry.area
perimeter = places.geometry.length

places["Circularity"] = (4 * np.pi * area) / (perimeter**2)
places_c = places.sort_values("Circularity" , ascending = False)

print(places_c[["NAME","Circularity"]].head(20))

```
